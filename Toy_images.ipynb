```python
import numpy as np
from matplotlib import pyplot as plt
```

Generate toy images

```python
s_image = np.array(([0]*4 + [1]*4)*4 + ([1]*4 + [0]*4)*4).reshape(8, 8)
plt.imshow(s_image, cmap='gray')
```

```python
s_image = np.array(((([0]*2 + [1]*2 + [0]*2 + [1]*2)*2) +
                    (([1]*2 + [0]*2 + [1]*2 + [0]*2)*2))*2
                    ).reshape(8, 8)
plt.imshow(s_image, cmap='gray')
```

```python
s_image = np.array(([0,1]*4 + [1,0]*4)*4).reshape(8, 8)
plt.imshow(s_image, cmap='gray')
```

```python
s_image = np.array([0,1] * 32).reshape((8, 8))
plt.imshow(s_image, cmap='gray')
```

```python
s_image = np.array([0,0,1,1] * 16).reshape((8, 8))
plt.imshow(s_image, cmap='gray')
```

```python
s_image = np.array([0,0,0,0,1,1,1,1] * 8).reshape((8, 8))
plt.imshow(s_image, cmap='gray')
```
